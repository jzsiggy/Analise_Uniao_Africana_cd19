# Análise de dados

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

## Indicador de IDH: 
### Índice de Desenvolvimento Humano, leva em conta vários aspectos do desenvolvimento de um país como educação, PIB e expectativa de vida para criar um índice que reflita  o desenvolvimento do país.

## Mortalidade Infantil:
### Indica quantas crianças não chegam aos seus 5 anos die idade a cada 1000 crianças.

## PIB per Capita:
### O PIB de certo país, dividido pelo número de seus habitantes.

## Fertilidade:
### Um indicador que mostra quantos filhos nascem por mãe. uma fertilidade muito alta geralmente está atrelada a um IDH menor, como demonstraremos em seguida.

## Porcentagem do PIB oriundo de Agricultura
### Qual porcentagem da riqueza de um país vem de Agricultura.

```python
IDH = pd.read_excel("Indicator_HDI.xlsx") # indicador de idh
child_mortality = pd.read_excel("indicator gapminder under5mortality.xlsx") # indicador de mortalidade infantil
pib_per_capita = pd.read_excel("indicator gapminder gdp_per_capita_ppp.xlsx") # indicador pib_per_capita
filhos_por_mae = pd.read_excel("indicator undata total_fertility.xlsx") # indicador de fertilidade
agricultura = pd.read_excel("agriculture_percent_of_gdp.xlsx") # indicador de porcentagem do pib fruto de agricultura
```

```python
# reindexando 

IDH = IDH.set_index("HDI")
filhos_por_mae = filhos_por_mae.set_index("Total fertility rate")
pib_per_capita = pib_per_capita.set_index("GDP per capita")
child_mortality = child_mortality.set_index("Under five mortality")
agricultura = agricultura.set_index("country")
```


    O segundo método de analise foi criar alguns dataframes com as colunas sendo os indicadores
    e as fileiras sendo os países. Desta forma, para cada indicador, poderemos vizualizar todos países.
    
    teremos varios dataframes, um para cada ano que queremos analizar.


```python
def make_df_by_year(year):
    paises = ["Sudan", "Egypt", "Cameroon", "Angola", "South Africa", "Libya", "Namibia", 
              "Zimbabwe", "Senegal", "Morocco", "Haiti", "Dominican Republic", "Afghanistan"]
    anos = [year]
    df = pd.DataFrame()
    df["Countries"] = paises
    df = df.set_index("Countries")
    df["mortalidade infantil"] = child_mortality[year]
    df = df.join(pib_per_capita[year])
    df.rename(columns={year :'pib per capita'}, inplace=True)
    df = df.join(filhos_por_mae[year])
    df.rename(columns={year :'Fertilidade'}, inplace=True)
    df = df.join(agricultura[year])
    df.rename(columns={year :'Agriculura (% pib)'}, inplace=True)
    df["Pertence a UA"] = 0
    df.loc[["Haiti", "Afghanistan", "Dominican Republic"], "Pertence a UA"] = -1
    return df
    
```

```python
paises_pertencentes_1965 = ["Libya", "Egypt", "Cameroon", "Senegal"]
paises_pertencentes_1970 = ["Libya", "Egypt", "Cameroon", "Senegal"]
paises_pertencentes_1975 = ["Libya", "Egypt", "Cameroon", "Senegal"]
paises_pertencentes_1980 = ["Libya", "Egypt", "Cameroon", "Angola", "Senegal"]
paises_pertencentes_1985 = ["Libya", "Egypt", "Cameroon", "Angola", "Zimbabwe", "Senegal"]
paises_pertencentes_1990 = ["Libya", "Egypt", "Cameroon", "Angola", "Zimbabwe", "Senegal"]
paises_pertencentes_1995 = ["Libya", "Egypt", "Cameroon", "Angola", "South Africa", "Namibia", "Zimbabwe", "Senegal"]
paises_pertencentes_2015 = ["Libya", "Egypt", "Cameroon", "Angola", "South Africa", "Sudan", "Namibia", "Zimbabwe", "Senegal", "Morocco"]
```

```python
Ano_1960 = make_df_by_year(1960)
Ano_1965 = make_df_by_year(1965)
Ano_1965.loc[paises_pertencentes_1965, "Pertence a UA"] = 1
Ano_1970 = make_df_by_year(1970)
Ano_1970.loc[paises_pertencentes_1970, "Pertence a UA"] = 1
Ano_1975 = make_df_by_year(1975)
Ano_1975.loc[paises_pertencentes_1975, "Pertence a UA"] = 1
Ano_1980 = make_df_by_year(1980)
Ano_1980.loc[paises_pertencentes_1980, "Pertence a UA"] = 1
Ano_1985 = make_df_by_year(1985)
Ano_1985.loc[paises_pertencentes_1985, "Pertence a UA"] = 1
Ano_1990 = make_df_by_year(1990)
Ano_1990.loc[paises_pertencentes_1990, "Pertence a UA"] = 1
Ano_1995 = make_df_by_year(1995)
Ano_1995.loc[paises_pertencentes_1995, "Pertence a UA"] = 1
Ano_2015 = make_df_by_year(2015)
Ano_2015.loc[paises_pertencentes_2015, "Pertence a UA"] = 1
```

A partir do gráfico abaixo, mostramos como o pib per capita dos países mudou ao longo dos últimos anos. Assim, sabendo os anos em que os países entraram para UA, podemos saber se a aderência ao bloco teve um forte impacto na mudança de seu pib per capita.

```python
anos = np.arange(1970, 2015, 1)
paises = ["Sudan", "Egypt", "Cameroon", "Angola", "South Africa", "Libya", "Namibia", 
              "Zimbabwe", "Senegal", "Morocco", "Haiti", "Dominican Republic", "Afghanistan"]

plt.figure(figsize=(10,10))
for ano in anos:
    for country in paises:
        if country in paises_pertencentes_2015:
            plt.scatter(ano, pib_per_capita.loc[country, ano], c="orange")
        if country not in paises_pertencentes_2015:
            plt.scatter(ano, pib_per_capita.loc[country, ano], c="blue")
            
plt.show()
```

Os histogramas abaixo nos permitem analisar no mesmo gráfico a Fertilidade de países na UA comparada com a de países fora da UA. Como demonstraremos abaixo, Esse é um indicador negativo.

```python
data_UA = filhos_por_mae.loc[paises_pertencentes_2015, :]
data_row = filhos_por_mae.loc[:, :]


data_UA[1970].plot.density()
data_row[1970].plot.density()

plt.show()


data_UA[1980].plot.density()
data_row[1980].plot.density()

plt.show()


data_UA[1990].plot.density()
data_row[1990].plot.density()

plt.show()


data_UA[2000].plot.density()
data_row[2000].plot.density()

plt.show()

```

Esses histogramas mostram uma tendência dos países da União Africana de diminuir a fertilidade, mas ainda está acima da tendência mundial.
