
    O segundo método de analise foi criar alguns dataframes com as colunas sendo os indicadores
    e as fileiras sendo os países. Desta forma, para cada indicador, poderemos vizualizar todos países.
    
    teremos varios dataframes, um para cada ano que queremos analizar.


```python
def make_df_by_year(year):
    paises = ["Sudan", "Egypt", "Cameroon", "Angola", "South Africa", "Libya", "Namibia", 
              "Zimbabwe", "Senegal", "Morocco", "Haiti", "Dominican Republic", "Afghanistan"]
    anos = [year]
    df = pd.DataFrame()
    df["Countries"] = paises
    df = df.set_index("Countries")
    df["mortalidade infantil"] = child_mortality[year]
    df = df.join(pib_per_capita[year])
    df.rename(columns={year :'pib per capita'}, inplace=True)
    df = df.join(filhos_por_mae[year])
    df.rename(columns={year :'Fertilidade'}, inplace=True)
    df = df.join(agricultura[year])
    df.rename(columns={year :'Agriculura (% pib)'}, inplace=True)
    df["Pertence a UA"] = 0
    df.loc[["Haiti", "Afghanistan", "Dominican Republic"], "Pertence a UA"] = -1
    return df
    
```

```python
paises_pertencentes_1965 = ["Libya", "Egypt", "Cameroon", "Senegal"]
paises_pertencentes_1970 = ["Libya", "Egypt", "Cameroon", "Senegal"]
paises_pertencentes_1975 = ["Libya", "Egypt", "Cameroon", "Senegal"]
paises_pertencentes_1980 = ["Libya", "Egypt", "Cameroon", "Angola", "Senegal"]
paises_pertencentes_1985 = ["Libya", "Egypt", "Cameroon", "Angola", "Zimbabwe", "Senegal"]
paises_pertencentes_1990 = ["Libya", "Egypt", "Cameroon", "Angola", "Zimbabwe", "Senegal"]
paises_pertencentes_1995 = ["Libya", "Egypt", "Cameroon", "Angola", "South Africa", "Namibia", "Zimbabwe", "Senegal"]
paises_pertencentes_2015 = ["Libya", "Egypt", "Cameroon", "Angola", "South Africa", "Sudan", "Namibia", "Zimbabwe", "Senegal", "Morocco"]
```

```python
Ano_1960 = make_df_by_year(1960)
Ano_1965 = make_df_by_year(1965)
Ano_1965.loc[paises_pertencentes_1965, "Pertence a UA"] = 1
Ano_1970 = make_df_by_year(1970)
Ano_1970.loc[paises_pertencentes_1970, "Pertence a UA"] = 1
Ano_1975 = make_df_by_year(1975)
Ano_1975.loc[paises_pertencentes_1975, "Pertence a UA"] = 1
Ano_1980 = make_df_by_year(1980)
Ano_1980.loc[paises_pertencentes_1980, "Pertence a UA"] = 1
Ano_1985 = make_df_by_year(1985)
Ano_1985.loc[paises_pertencentes_1985, "Pertence a UA"] = 1
Ano_1990 = make_df_by_year(1990)
Ano_1990.loc[paises_pertencentes_1990, "Pertence a UA"] = 1
Ano_1995 = make_df_by_year(1995)
Ano_1995.loc[paises_pertencentes_1995, "Pertence a UA"] = 1
Ano_2015 = make_df_by_year(2015)
Ano_2015.loc[paises_pertencentes_2015, "Pertence a UA"] = 1
```
